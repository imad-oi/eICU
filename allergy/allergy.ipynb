```python
import pandas as pd

# Charger les données
url = "https://physionet.org/files/eicu-crd-demo/2.0.1/"
allergy_data = pd.read_csv(url + "allergy.csv.gz")

allergy_data.head()
```

```python
# 2. Supprimez les colonnes qui ne sont pas utiles.
colonnes_inutiles = ["allergyid","patientunitstayid"]

allergy_data = allergy_data.drop(colonnes_inutiles, axis=1)
```

```python
# 3. Supprimez les lignes avec des valeurs manquantes.
allergy_data = allergy_data.dropna()
```

```python
# 4. Prenez soin des données manquantes.
columns_to_interpolate = ["allergyoffset", "allergyenteredoffset", "drughiclseqno"]

for column in columns_to_interpolate:
    allergy_data[column] = allergy_data[column].interpolate()
    
columns_to_fill = ["allergynotetype","specialtytype","usertype"	,"allergyname", "allergytype", "drugname"]

for column in columns_to_fill:
    allergy_data[column] = allergy_data[column].fillna("Unknown")
```
