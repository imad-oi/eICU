```python
import pandas as pd

# Charger les données
url = "https://physionet.org/files/eicu-crd-demo/2.0.1/"
apachPredVar_data = pd.read_csv(url + "apachePredVar.csv.gz")

apachPredVar_data.head()
```

```python
# 2. Supprimez les colonnes qui ne sont pas utiles.
colonnes_inutiles = ["apachepredvarid", "patientunitstayid"]
apachPredVar_data = apachPredVar_data.drop(columns=colonnes_inutiles)
```

```python
# 3. Supprimez les lignes avec des valeurs manquantes.
apachPredVar_data = apachPredVar_data.dropna()
```

```python
apachPredVar_data.head()
```

```python
# 4. Prenez soin des données manquantes.

columns_to_fill = ["admitdiagnosis"]
for column in columns_to_fill:
    apachPredVar_data[column] = apachPredVar_data[column].fillna("unknown")
    
# for numbers we can use the mean
# exclude the columns that are not numbers
columns_of_type_numbers = [col for col in apachPredVar_data.columns if col not in columns_to_fill]

print(columns_of_type_numbers)

# fill the missing values with the mean
apachPredVar_data[columns_of_type_numbers] = apachPredVar_data[columns_of_type_numbers].fillna(apachPredVar_data[columns_of_type_numbers].mean())

```

```python
# 5. Convertissez le bloc de données en NumPy.
import numpy as np
apachPredVar_data = apachPredVar_data.to_numpy()
print(apachPredVar_data)
```
