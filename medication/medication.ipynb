```python
import pandas as pd

# Charger les données
url =   "https://physionet.org/files/eicu-crd-demo/2.0.1/"
medication_data = pd.read_csv(url + "medication.csv.gz")

medication_data.head()
```

```python
medication_data.columns
```

```python
# Supprimez les colonnes qui ne sont pas utiles
colonnes_inutiles = ['medicationid',      # car on a déjà un index
                     'patientunitstayid', # car on a déjà un index
                     'drugorderoffset',   
                     'drugstartoffset',
                     'drugivadmixture', 
                     'drugordercancelled',
                     'drughiclseqno', 
                     'drugstopoffset',
                     'gtc',
                     'loadingdose']
medication_data.drop(columns=colonnes_inutiles, inplace=True)
```

```python
# drop rows with missing values
medication_data.dropna()
```

```python
# prenez sois des valeurs manquantes
# 1- les nombres manquants pour la colonne dosage par la moyenne
# calculer la moyenne
medication_data['dosage'] = medication_data['dosage'].interpolate()

# 2- les string manquants
medication_data.fillna('Unknown').head()
```
